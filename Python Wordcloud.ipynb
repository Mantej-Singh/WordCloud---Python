```python
# Generate a word cloud 
wordcloud = WordCloud().generate(WordCloud_data)
```

```python
# lower max_font_size
wordcloud = WordCloud(font_path='C:/Users/Jarvis/ARegular.ttf',background_color='black',mode='RGB',width=2000,height=1000).generate(WordCloud_data)
#you can add fonts, stopwords, background color, adjust hight width, and many other options
plt.title("Samsung - issues with Galaxy Note 7")
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

# Trying to make Wc with random colors

```python
import random
```

```python
def random_color(word, font_size, position, orientation, random_state=None, **kwargs):
    return "#%06x" % random.randint(0, 0xFFFFFF)
```

```python
wordcloud = WordCloud(background_color='black').generate(WordCloud_data)
plt.imshow(wordcloud.recolor(color_func=random_color))
plt.axis("off")
plt.show()
```
